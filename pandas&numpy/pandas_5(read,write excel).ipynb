```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'sam walton'
    return cell

def convert_esp_cell(cell):
    return None if cell=='not available' else cell


df=pd.read_excel("C:\\Users\\Dell\\Desktop\\python\\stock_data.xlsx"
                 ,"sheet1",converters={
                     'people':convert_people_cell,
                     'eps':convert_esp_cell,
                    'revenue': lambda x: 0 if x == 50 else x
                 })
df
```

```python
df.to_excel("C:\\Users\\Dell\\Desktop\\python\\stock_data(1).xlsx",sheet_name="sheet1",index=False,header=False)
```

```python
df.to_excel("stock_data.xlsx",sheet_name="sheet1",startrow=1,startcol=2,index=False,header=False)
# FOR DEMO

```
