```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=["GOOGL","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.bar(ypos,revenue,label="revenue")


plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")

plt.legend()
```

```python
company=["GOOGL","AMZN","MSFT","FB"]
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.bar(xpos-0.2,revenue,width=0.45,label="revenue")
plt.bar(xpos+0.2,profit,width=0.35,label="profit")

plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")

plt.legend()
```

```python
plt.yticks(ypos,company)
plt.barh(xpos-0.2,revenue,label="revenue")
plt.barh(xpos+0.2,profit,label="profit")

plt.xlabel("revenue(bin)")
plt.title("US Tech Stocks")

plt.legend()
```
