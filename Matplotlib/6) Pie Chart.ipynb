```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.2f%%')
# plt.axis("equal")
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.2f%%',explode=[0,0.2,0,0.2,0])
plt.show()
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct='%0.2f%%',
        explode=[0,0.2,0,0.2,0],startangle=180)
plt.show()
```
